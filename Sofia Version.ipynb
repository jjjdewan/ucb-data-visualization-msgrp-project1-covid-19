```python
# Dependencies
import csv
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress 
import seaborn as sns 
from scipy import stats
import matplotlib.dates as mdates
import scipy.stats as st
from pprint import pprint
from datetime import datetime
from geopy.exc import GeocoderTimedOut 
from geopy.geocoders import Nominatim
import gmaps
from api_keys import g_key
gmaps.configure(api_key=g_key)
```

```python
# Import the covid-19 dataset 
csv_path = ('Resources/owid-covid-data_updated.csv')
covid_data_df = pd.read_csv(csv_path)
covid_data_df
```

```python
#Total Cases, Deaths and Mortality by Continent
cases_by_continent = covid_data_df.groupby(["continent"]).agg({'total_cases':['sum'],'total_deaths':['sum']})

cases_by_continent["Mortality_rate"]= round((cases_by_continent['total_deaths']/ cases_by_continent['total_cases'])*100,2)

cases_by_continent.columns=['Total Cases','Total Deaths', 'Mortality Rate in %']
cases_by_continent
```

```python
#Top 10 Countries with highest mortality rate
cases_by_country = covid_data_df.groupby(["location"]).agg({'total_cases':['sum'],'total_deaths':['sum']})

cases_by_country["Mortality_rate"]= round((cases_by_country['total_deaths']/ cases_by_country['total_cases'])*100,2)

cases_by_country.columns=['Total Cases','Total Deaths', 'Mortality Rate in %']
top=cases_by_country.sort_values('Mortality Rate in %',ascending=False)
top.head(10)
```

```python
#Average life expectacy for each continent
life_expectancy_by_continent = covid_data_df.groupby(["continent"])['life_expectancy'].mean()

#Average GDP for each continent
avg_gdp_continent=covid_data_df.groupby(['continent'])['gdp_per_capita'].mean()

#Mean Population sensity per continent
pop_density=covid_data_df.groupby(['continent'])['population_density'].mean()

Summary_df=pd.DataFrame({ 'Life Expectancy':life_expectancy_by_continent, 'GDP':avg_gdp_continent, 'Population Density':pop_density})
Summary_df
```

```python
#Merged dataframe

summarized_df=pd.merge(cases_by_continent, Summary_df, how="left", on="continent")
summarized_df
```

```python
x_values = summarized_df['Life Expectancy']
y_values = summarized_df['Mortality Rate in %']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
#plt.annotate(line_eq,(7,10),fontsize=10,color="black")
plt.xlabel('Life Expectancy')
plt.ylabel('Mortality Rate')
plt.show()
```

```python
x_values = summarized_df['GDP']
y_values = summarized_df['Mortality Rate in %']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(10,1),fontsize=10,color="black")
plt.xlabel('GDP')
plt.ylabel('Mortality Rate')
plt.show()
```

```python
x_values = summarized_df['Population Density']
y_values = summarized_df['Total Cases']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=10,color="black")
plt.xlabel('Population Density')
plt.ylabel('Total Cases')
plt.show()
```

```python
#Feed the regression equation to get expected Covid cases
summarized_df['Expected Covid Cases based on Population Density']=(summarized_df['Population Density'])*4989.47 + 4161587.74
summarized_df
```

```python
# With four rows, the degree of freedom is 6-1 = 5
# With a p-value of 0.05, the confidence level is 1.00-0.05 = 0.95.
critical_value = stats.chi2.ppf(q = 0.95, df = 5)
```

```python
# Run the chi square test with stats.chisquare()
stats.chisquare(summarized_df['Total Cases'], summarized_df['Expected Covid Cases based on Population Density'])
```

```python
longitude = [] 
latitude = [] 
   
# function to find the coordinate 
# of a given country  
def findGeocode(country): 
       
    # try and catch is used to overcome 
    # the exception thrown by geolocator 
    # using geocodertimedout   
    try: 
          
        # Specify the user_agent as your 
        # app name it should not be none 
        geolocator = Nominatim(user_agent="your_app_name") 
          
        return geolocator.geocode(country) 
      
    except GeocoderTimedOut: 
          
        return findGeocode(country)
```

```python
for i in (covid_data_df["location"]): 
      
    if findGeocode(i) != None: 
           
        loc = findGeocode(i) 
          
        # coordinates returned from  
        # function is stored into 
        # two separate list 
        latitude.append(loc.latitude) 
        longitude.append(loc.longitude) 
       
    # if coordinate for a city not 
    # found, insert "NaN" indicating  
    # missing value  
    else: 
        latitude.append(np.nan) 
        longitude.append(np.nan) 
```

```python
covid_data_df["Latitude"] = latitude
covid_data_df["Longitude"] = longitude 
```
