```python
# Dependencies
import csv
import pandas as pd
import os
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress 
import seaborn as sns 
from scipy import stats
import matplotlib.dates as mdates
import scipy.stats as st
from pprint import pprint
from datetime import datetime
```

```python
# Import the covid-19 dataset 
csv_path = ('Resources/owid-covid-data_updated_main.csv')
covid_data_df = pd.read_csv(csv_path)
covid_data_df.head()
```

```python
covid_data_df.dropna()
covid_data_df
```

```python
def linear_regress(x,y):
    print(f"The r-squared is : {round(st.pearsonr(x, y)[0],2)}")
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
    regress_values = x * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(x, y)
    plt.plot(x,regress_values,"r-")
    return line_eq

# Fuction for annotating plot
#
def annotate_plt(line_eq, a, b):
    plt.annotate(line_eq,(a,b),fontsize=15,color="blue")
```

```python
get_regress_life = linear_regress(covid_data_df["extreme_poverty"],
                             covid_data_df["total_deaths"]/100)

# Call annotate func
annotate_plt(get_regress_life, 30, 100)

# Set a title
plt.title("Covid-19 cases: hospital beds vs. total deaths")

# Set xlabel
plt.xlabel("Hospital Beds per 1000")

# Set ylabel
plt.ylabel("Total # of Deaths * 1000")

# Set grid 
plt.grid(linestyle='-', linewidth=1.5, alpha = 0.5, color='r')

# Save the figure
plt.savefig("Covid-19 cases: life expectancy average vs. total deaths.png")
```
